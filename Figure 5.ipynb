```julia
using Plots

# Parameters
nt = 15
nh = 10000

# --- Read binary file as Float32 and reshape ---
fid = open("data_fig5.bin", "r")
raw_data = read!(fid, Vector{Float32}(undef, nt*nh))  # read nt*nh floats
close(fid)

data_nmo = reshape(raw_data, nt, nh)  # nt rows × nh columns (like MATLAB)

# --- Time axis ---
dt_ms = 6   # MATLAB overwrote dt with 6
T = (0:nt-1) .* dt_ms

# --- Plotting ---
plot(; size=(700,300), legend=false,
      xlabel="Trace index", ylabel="Time (ms)", title="a)")

for i in 1:70:nh
    plot!(data_nmo[:, i] .* 200 .+ i, T; color=:black, linewidth=2)
end

xlims!(0, nh)
ylims!(0, T[end])
```
